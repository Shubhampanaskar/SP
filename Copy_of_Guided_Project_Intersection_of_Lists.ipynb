Given a list of n lists, write a function which returns the intersection of all the n lists. Each list can have any number of elements.

Example 1:

```
input = [[5, 5, 5, 8, 10], 
      [5, 5, 7, 8], 
      [5, 5, 8, 9, 11]]
```

```
output = [5, 5, 8]
```

Example 2:

```
input = [[5, 6], 
      [5, 5, 7, 8], 
      [5, 5, 8, 9]]
```


```
output = [5]
```

````python
input = [[5, 5, 5, 8, 10], 
      [5, 5, 7, 8], 
      [5, 5, 8, 9, 11]]
````

````python

def get_min_list_in_list_of_list(list_of_list):
 
    min_nu_l=list_of_list[0]
    min_nu_=len(list_of_list[0])
    for list1 in list_of_list:
      if len(list1) < min_nu_ :
        min_nu_=len(list1)
        min_nu_l= list1
    else:
      pass
    return min_nu_l      
  
````

````python
get_min_list_in_list_of_list(input)
````
